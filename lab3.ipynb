```python
# Don't edit this function
def load_sample():
    """Return the entire text of Hamlet in a string."""
    with open('hamlet.txt') as f:
        sample = f.read()
    return sample
```

```python
# Edit this function. "pass" tells Python to do nothing.
def parse_sample(sample, words_only, sort_list): 
    ss = sample.replace('\t', ' ')
    ss = ss.replace('\n', ' ')
    ss = ss.replace('-', ' ')
    list = ss.split(' ')
    list = [x for x in list if x != '']
    
    if words_only:
        punctuation = [',','.','/','"','?','"',']','[',';',':']
        length = len(list)
        for i in range(len(list)):
            for pun in punctuation:
                list[i] =  "".join(list[i].split(pun))
                list[i] = list[i].lower()
    if sort_list:
        list = sorted(list)
        
    return list
```

```python
parse_sample(load_sample(), True, True)
```

```python
def count_freq(lst):
    """Return a dict of word frequencies given a list of words.
    
    Arguments:
    -- lst: a list of strings
    
    Returns:
    -- freq_dict: a dictionary of word: frequency pairs. The keys are strings
       containing the words, and the values are integers indicating how many
       times each word appears.
    
    Example:
    >>> count_freq(['time', 'after', 'time'])
    {'time': 2, 'after': 1}
    """
    freq = []
    for w in lst:
        freq.append(lst.count(w))
    freq_dict = dict(zip(lst,freq))
    return freq_dict

def mean(lst):
    """Return the arithmetic mean of a list of numbers."""
    
    total =0
    for i in lst:
        total = total + i
    return total/len(lst)

def stdev(lst):
    """Return the standard deviation of a list of numbers."""
    m = mean(lst)
    lst2 = []
    for i in lst:
        lst2.append((i-m)**2)
    return mean(lst2)**(.5)

def median(lst):
    """Return the median of a list of numbers."""
    med = 0
    lst3 = sorted(lst)
    if len(lst3)%2 ==0:
        med = (lst3[(len(lst3)/2)]+lst[(len(lst3)/2)-1])/2.0
    else:
        med = lst3[(len(lst3)/2)]
    return med
```

```python
def print_stats(lst):
    print "Mean: ", mean(lst)
    print "Median: ", median(lst)
    print "Standard Deviation: ", stdev(lst)
```
